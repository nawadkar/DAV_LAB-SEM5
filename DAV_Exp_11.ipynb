## **Data Visualization Libraries in Python**

### **Matplotlib:**

**Features:**
*   Versatile 2D plotting library for creating static visualizations.
*   Provides a wide range of plot types, including line plots, scatter plots, bar plots, and more.

**Applications:**
*   Basic and intermediate-level data visualization for understanding and presenting data in various formats.



### **Seaborn:**

**Features:**
*   Built on Matplotlib, offering a high-level interface for statistical graphics.
*   Simplifies the creation of aesthetically pleasing and informative visualizations.

**Applications:**
*   Statistical data visualization with a focus on concise and attractive representation of complex relationships.



### **Pandas Plotting:**

**Features:**
*   Integrated plotting interface in Pandas for quick visualization from DataFrames.
*   Supports various plot types, making it convenient for exploratory data analysis.

**Applications:**
*   Rapid and easy plotting of data directly from Pandas DataFrames during data analysis tasks.

*   Enables the creation of interactive and shareable visualizations, suitable for web applications.

**Applications:**
*   Building interactive and dynamic charts for online presentations, dashboards, and web-based data exploration.


### **Bokeh:**

**Features:**
*   Interactive visualization library with a focus on creating web-ready, interactive plots.
*   Designed for the development of interactive dashboards and applications with real-time updates.

**Applications:**
*   Developing web-based applications and dashboards that require dynamic, interactive visualizations for data exploration.
*   These libraries cater to different needs, ranging from simple static plots to complex, interactive visualizations, making them suitable for various data analysis and presentation scenarios.

## **Python**

### **MatplotLib**

```r
import matplotlib.pyplot as plt

categories = ['AIDS', 'EXTC', 'CMPN', 'IT']
values = [30, 20, 15, 35]

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999'])
plt.title('Pie Chart with Matplotlib')

plt.show()
```

### **Seaborn**

```r
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)

sns.scatterplot(x=x, y=y, label='Scatter Plot', color='blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot with Seaborn')

plt.legend()
plt.show()
```

## **R**

### **ggplot2**

```r
library(ggplot2)

data <- data.frame(
  category = c('AIDS', 'EXTC', 'CMPN', 'IT'),
  value = c(30, 20, 15, 35)
)

ggplot(data, aes(x = "", y = value, fill = category)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  labs(title = "Pie Chart with ggplot2", fill = "Categories") +
  theme_void()
```

### **Lattice**

```r
library(lattice)

# Generate random data
set.seed(42)
x <- rnorm(50)
y <- 2 * x + 1 + 0.1 * rnorm(50)

# Create a scatter plot using lattice
xyplot(y ~ x, data = data.frame(x, y), type = 'p', col = 'blue',
       xlab = 'X-axis', ylab = 'Y-axis', main = 'Simple Scatter Plot with lattice')
```
