```r
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(data)

# Shapiro-Wilk test for normality
stat, p_value = stats.shapiro(df['Year'])
print(f'Shapiro-Wilk Test Statistic: {stat}, p-value: {p_value}')

# Q-Q plot
plt.figure(figsize=(8, 4))
sns.set(style="whitegrid")
stats.probplot(df['Year'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Year')
plt.show()
```

```r
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming df is your DataFrame
year_data = df['Year']
GDP_In_Billion_USD_data = df['GDP_In_Billion_USD']

# Reshape the data for scikit-learn
X = np.array(year_data).reshape(-1, 1)
y = np.array(GDP_In_Billion_USD_data)

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Print the coefficients and intercept
print(f"Coefficient: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")

# Predictions for the entire 'Year' column
predictions = model.predict(X)

# Plot the regression line on the scatter plot
plt.scatter(year_data,GDP_In_Billion_USD_data, label='Actual Data')
plt.plot(year_data, predictions, color='red', label='Linear Regression')
plt.title('Linear Regression: Year vs GDP_In_Billion_USD ')
plt.xlabel('Year')
plt.ylabel('GDP_In_Billion_USD')
plt.legend()
plt.show()
```
